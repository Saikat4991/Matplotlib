```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/ne_India_rainfall_act_dep_1901_2015.csv')  # Replace 'your_dataset.csv' with your dataset file name
df = df.iloc[:,:11]  # Display the first few rows of the dataset
df.head()
```

```python
df.columns
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['YEAR'], df['Actual Rainfall: JUN'], label='June Rainfall')
plt.scatter(df['YEAR'], df['Actual Rainfall: JUL'], label='July Rainfall')
plt.scatter(df['YEAR'], df['Actual Rainfall: AUG'], label='August Rainfall')
plt.scatter(df['YEAR'], df['Actual Rainfall: SEPT'], label='September Rainfall')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.title('Monthly Rainfall Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([df['Actual Rainfall: JUN'], df['Actual Rainfall: JUL'], df['Actual Rainfall: AUG'], df['Actual Rainfall: SEPT']])
plt.xticks([1, 2, 3, 4], ['June', 'July', 'August', 'September'])
plt.ylabel('Rainfall (mm)')
plt.title('Monthly Rainfall Distribution')
plt.grid(True)
plt.show()
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(df['YEAR'], df['Actual Rainfall: JUN-SEPT'], label='Total Rainfall', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Rainfall (mm)')
plt.title('Total Rainfall During June-September')
plt.legend()
plt.grid(True)
plt.show()
```
