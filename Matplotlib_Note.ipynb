```python
## Repeat data
## Histogram : graph frequency
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize = (6,3))
plt.hist(data)
```

```python
plt.figure(figsize = (6,3))
plt.hist(data, color = "#700EBE" )
plt.xlabel("Representing Unique Values")
plt.ylabel("Representing Corresponding Frequencies")
plt.title("Histogram")
plt.grid()
plt.savefig('Histogram')
```

```python
## 3-D graph
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c = '#9327E8', alpha = .9)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# plt.zlabel('Z-axis')
plt.show()
```
